```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Random Module

```python
# Random floats
random_floats = np.random.rand() # random floats between 0 to 1
print(random_floats)

# Random int
random_int = np.random.randint(1, 10) # random number between 1 to 9
print(random_int)

# Random arrays
random_array_floats = np.random.rand(3, 2) # 3*2 array of random floats
print(random_array_floats)

# Random arrays
random_array_int = np.random.randint(1, 10, size=(3, 2)) # 3*2 array of int
print(random_array_int)

```

### Random Seed

```python
# To ensure reproducibility, you can set a random seed. This will ensure that the random numbers generated are the same each time you run the code.
# Setting the seed ensures that the random number sequence is repeatable.
np.random.seed(42)
random_once = np.random.rand()
print(random_once)
```

### Random Samples

```python
# Use numpy.random.choice() to generate random samples from a given list or array.
sample = np.random.choice([1, 2, 3, 4, 5], size=3)
print(sample)

# By default, choice samples with replacement. To sample without replacement, set replace=False.
sample_without_replacement = np.random.choice([1, 2, 3, 4, 5], size=3, replace=False) # No Duplicates
print(sample_without_replacement) 
```

### Shuffling Arrays

```python
# You can shuffle the elements in an array using numpy.random.shuffle() which modifies the array in place.
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)

# For a shuffled version of the array without modifying the original, use numpy.random.permutation().
arr = np.array([1, 2, 3, 4, 5])
shuffled_arr = np.random.permutation(arr)
print(shuffled_arr)
```

## Data Distribution in NumPy

### Uniform Distribution
The uniform distribution is a type of continuous probability distribution where all outcomes are equally likely within a specified range.  

Importance in Data Science:
1. Used to generate random samples for simulations.
2. Acts as a baseline for comparing other distributions.

```python
# Generate 1000 samples
uniform_samples = np.random.uniform(0, 1, 1000)

# Plot
sns.histplot(uniform_samples, bins=30, kde=True)
plt.title("Uniform Distribution")
plt.show()
```

### Normal (Gaussian) Distribution
The normal distribution is a continuous probability distribution characterized by a bell-shaped curve.  
Defined by its mean (μ) and standard deviation (σ).  

Importance in Data Science:
1. Many statistical tests assume normality.
2. Central Limit Theorem: sums of independent random variables tend to be normally distributed.
3. Used in confidence intervals and hypothesis testing.

```python
# Generate 1000 samples
normal_samples = np.random.normal(0, 1, 1000)

# Plot
sns.histplot(normal_samples, bins=30, kde=True)
plt.title("Normal Distribution")
plt.show()
```

### Binomial Distribution
The binomial distribution is a discrete probability distribution.  
Represents the number of successes in a sequence of n independent experiments.  

Importance in Data Science:
1. Used in binary classification problems.
2. Model scenarios with two possible outcomes (e.g., success/failure).

```python
# Generate 1000 samples
binomial_samples = np.random.binomial(10, 0.5, 1000)

# Plot
sns.histplot(binomial_samples, bins=30, kde=True)
plt.title("Binomial Distribution")
plt.show()
```

### Poisson Distribution
The Poisson distribution is a discrete probability distribution.  
Represents the number of events occurring within a fixed interval.  

Importance in Data Science:
1. Used for modeling count data.
2. Applied in fields like traffic flow, call centers, and natural events.

```python
# Generate 1000 samples
poisson_samples = np.random.poisson(3, 1000)

# Plot
sns.histplot(poisson_samples, bins=30, kde=True)
plt.title("Poisson Distribution")
plt.show()
```

### Exponential Distribution
The exponential distribution is a continuous probability distribution.  
Describes the time between events in a Poisson process.  

Importance in Data Science:
1. Models time until an event occurs.
2. Used in survival analysis and reliability engineering.

```python
# Generate 1000 samples
exponential_samples = np.random.exponential(1, 1000)

# Plot
sns.histplot(exponential_samples, bins=30, kde=True)
plt.title("Exponential Distribution")
plt.show()
```

### Chi-Square Distribution
The chi-square distribution is a continuous probability distribution.  
Represents the sum of the squares of k independent standard normal random variables.  

Importance in Data Science:
1. Used in hypothesis testing (e.g., Chi-square test for independence).
2. Construction of confidence intervals for variance.

```python
# Generate 1000 samples
chi_square_samples = np.random.chisquare(2, 1000)

# Plot
sns.histplot(chi_square_samples, bins=30, kde=True)
plt.title("Chi-Square Distribution")
plt.show()
```

### Logistic Distribution
The logistic distribution is similar to the normal distribution but has heavier tails.  

Importance in Data Science:
1. Used in logistic regression.
2. Models growth processes and logistic population growth.

```python
# Generate 1000 samples
logistic_samples = np.random.logistic(0, 1, 1000)

# Plot
sns.histplot(logistic_samples, bins=30, kde=True)
plt.title("Logistic Distribution")
plt.show()
```

### Multinomial Distribution
The multinomial distribution generalizes the binomial distribution for experiments with more than two possible outcomes.  

Importance in Data Science:
1. Used in natural language processing for word frequencies.
2. Multiclass classification problems.

```python
# Generate 1000 samples
multinomial_samples = np.random.multinomial(20, [0.1, 0.2, 0.3, 0.4], 1000)
multinomial_samples_sum = np.sum(multinomial_samples, axis=0)

# Plot
plt.bar(range(len(multinomial_samples_sum)), multinomial_samples_sum)
plt.title("Multinomial Distribution")
plt.show()
```
